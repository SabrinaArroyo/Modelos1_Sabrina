```python
# Cancer 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = r"/Users/sabrina/Documents/UNI/A-NEUROCIENCIAS/QUINTO /MODELOS/Neurociencias-2025-1/repoprofe/datasets/cancer_regression/cancer_reg.csv"
# la r antes para que no lea los caracteres especiales

df_cancer = pd.read_csv(csv_path, sep= ",")
```

```python
filtrado_col1 = df_cancer['pctsomecol18_24']
media = filtrado_col1.mean()
desviacion_estandar = filtrado_col1.std()

plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label=f'Media: {media:.2f}')
plt.axvline(media + desviacion_estandar, color='green', linestyle='dashed', linewidth=1, label=f'+1 Sigma: {media + desviacion_estandar:.2f}')
plt.axvline(media - desviacion_estandar, color='green', linestyle='dashed', linewidth=1, label=f'-1 Sigma: {media - desviacion_estandar:.2f}')

plt.hist(df_cancer['pctsomecol18_24'], bins=30)
plt.title('Histograma de pctsomecol18_24')
plt.xlabel('pctsomecol18_24')
plt.ylabel('Frecuencia')
filtrado_col1.describe()
```

```python
filtrado_col2 = df_cancer['pctemployed16_over']
media2 = filtrado_col2.mean()
desviacion_estandar2 = filtrado_col2.std()
plt.axvline(media2, color='red', linestyle='dashed', linewidth=1, label=f'Media: {media2:.2f}')
plt.axvline(media2 + desviacion_estandar2, color='green', linestyle='dashed', linewidth=1, label=f'+1 Sigma: {media2 + desviacion_estandar2:.2f}')
plt.axvline(media2 - desviacion_estandar2, color='green', linestyle='dashed', linewidth=1, label=f'-1 Sigma: {media2 - desviacion_estandar2:.2f}')

plt.hist(df_cancer['pctemployed16_over'], bins=30)
plt.title('Histograma de pctemployed16_over')
plt.xlabel('pctemployed16_over')
plt.ylabel('Frecuencia')
filtrado_col2.describe()
```

```python
filtrado_col3 = df_cancer['pctprivatecoveragealone']
media = filtrado_col3.mean()
desviacion_estandar = filtrado_col3.std()
plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label=f'Media: {media:.2f}')
plt.axvline(media + desviacion_estandar, color='green', linestyle='dashed', linewidth=1, label=f'+1 Sigma: {media + desviacion_estandar:.2f}')
plt.axvline(media - desviacion_estandar, color='green', linestyle='dashed', linewidth=1, label=f'-1 Sigma: {media - desviacion_estandar:.2f}')

plt.hist(df_cancer['pctprivatecoveragealone'], bins=30)
plt.title('Histograma de pctprivatecoveragealone')
plt.xlabel('pctprivatecoveragealone')
plt.ylabel('Frecuencia')
filtrado_col3.describe()
```

```python
df_cancer.plot.scatter(x='avganncount', y='avgdeathsperyear') # hay una correlacion muy evidente, entre + diagnostiquen + se mueren
x = df_cancer['avganncount']
y = df_cancer['avgdeathsperyear']

# Ajuste lineal
m, b = np.polyfit(x, y, 1)
# Agregar la lnea de correlacin
plt.plot(x, m * x + b, color='red', linestyle='dashed', linewidth=1, label='Lnea de Correlacin')
```
