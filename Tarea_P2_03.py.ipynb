```python
# Ejercicio 4. Genere 4 ondas sinusoidales con cada frecuencia y amplitud.
# Hint: recuerde que la ecuación de la onda sinusoidal es y(t) = A*sin(2*piB*t) donde A es la amplitud y B es la
# frecuencia. Para usar pi en numpy, use np.pi
# Sugerencia: para visualizar las ondas sinusoidales puede usar las líneas de código

import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de tiempo
t = np.linspace(0, 1, 1000)  # 1000 puntos entre 0 y 1 segundo

# Definir amplitudes y frecuencias para 4 ondas
amplitudes = [1, 2, 0.5, 1.5]
frecuencias = [5, 10, 2, 7]  # En Hz

# Crear una figura para visualizar las ondas
plt.figure(figsize=(10, 6))

# Generar y visualizar las 4 ondas
for i in range(4):
    A = amplitudes[i]
    B = frecuencias[i]
    y = A * np.sin(2 * np.pi * B * t)  # Ecuación de la onda sinusoidal
    plt.plot(t, y, label=f'Amplitud={A}, Frecuencia={B}Hz')

# Añadir etiquetas y leyenda
plt.title('Ondas Sinusoidales')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
# Ejercicio 5. Cree un arreglo llamado x, que sea la suma de las 4 ondas sinusoidales
amplitudes = [1, 2, 0.5, 1.5]
frecuencias = [5, 10, 2, 7] 
x = sum(A * np.sin(2 * np.pi * B * t) for A, B in zip(amplitudes, frecuencias))
x
```

```python
# Numpy también permite aplicar operadores tales como la transormada de fourier a un arreglo de datos con
# el paquete np.fft, en particular la función np.fft.fft(x) aplica la transformada discreta de fourier al arreglo x
X = np.fft.fft(x)
```
